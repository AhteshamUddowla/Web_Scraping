## Main

```python
def iHadis_scraping(url, book, chapter_bangla_name):
    chapter, bangla_name = chapter_bangla_name
    for i in range(1, chapter+1):
        new_url = f'{url}{book}/{i}'
        
        driver.get(new_url)
        time.sleep(20)
        
        hadis_no_elements = driver.find_elements(By.CSS_SELECTOR, 'div:nth-child(1) > div > div.flex.items-center.gap-x-3 > div.style-Kalpurush-webKit.text-xl.leading-6.text-primary')
        hadis_status_elements = driver.find_elements(By.CSS_SELECTOR, 'div > button > p')
        arabic_texts_elements = driver.find_elements(By.CSS_SELECTOR, 'div.bg-white.dark\:bg-hadith-dark-card.rounded-b-xl.p-6.xs\:py-1.md-max\:px-4 > div.arabic.dark\:text-hadith-deepoffwhite.w-full.text-right.mb-8.xss\:mt-4.xs\:mt-4.xs\:leading-\[40px\] > p')
        hadis_narrators_elements = driver.find_elements(By.CSS_SELECTOR, 'div.bg-white.dark\:bg-hadith-dark-card.rounded-b-xl.p-6.xs\:py-1.md-max\:px-4 > p.style-Kalpurush-primary.text-primary.leading-5.xss\:text-\[15px\].xss\:leading-7')
        bangla_texts_elements = driver.find_elements(By.CSS_SELECTOR, 'div.bg-white.dark\:bg-hadith-dark-card.rounded-b-xl.p-6.xs\:py-1.md-max\:px-4 > p.mt-4.style-Kalpurush.dark\:text-hadith-deepoffwhite.leading-10 > span')
        
  
        hadis_no = [i.text for i in hadis_no_elements]
        hadis_status = [i.text for i in hadis_status_elements[2::3]]
        arabic_texts = [i.text for i in arabic_texts_elements]
        hadis_narrators = [i.text for i in hadis_narrators_elements]
        bangla_texts = [i.text for i in bangla_texts_elements]
        
        book_name =bangla_name+'/'
        book_name = (book_name*len(hadis_no)).split('/')
        
        print(book, i, '-->', len(hadis_no), len(hadis_status), len(arabic_texts), len(hadis_narrators), len(bangla_texts))
        
        
        df_lst = [book_name, hadis_no, hadis_status, arabic_texts, hadis_narrators, bangla_texts]
        column_lst = ['বই', 'হাদিস নাম্বার', 'হাদিসের মান', 'আরবি', 'বর্ণনাকারী', 'বাংলা']
        df = pd.DataFrame(df_lst).T
        df.columns = column_lst
        df.to_csv(f'{book}/Chapter_{i}.csv', index=False)
    
```
