```python
session = HTMLSession()
```

```python
for i in range(22, 25):
    url = f'https://www.dsebd.org/old_news.php?startDate=20{i}-01-01&endDate=20{i}-12-31&criteria=4&archive=news'
    
    response = session.get(url)
    
    table_news = response.html.find('table.table-news',first=True)
    
    news_lst = (table_news.text).split('\n')
    
    trading_code = news_lst[1::8][:-4]
    news_title = news_lst[3::8][:-4]
    news = news_lst[5::8][:-4]
    post_date = news_lst[7::8][:-4]
    
    print(f'20{i} --> ', len(trading_code), len(news_title), len(news), len(post_date))
    
    df_news = [trading_code, news_title, news, post_date]
    columns_lst = ['Trading Code', 'News Title', 'News', 'Post Date']
    
    df = pd.DataFrame(df_news).T
    df.columns = columns_lst
    
    df.to_csv(f'20{i}.csv', index=False)
```
