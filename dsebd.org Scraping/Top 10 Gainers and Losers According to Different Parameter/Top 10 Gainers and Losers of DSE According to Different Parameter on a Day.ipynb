```python
import pandas as pd
from requests_html import HTMLSession
```

```python
session = HTMLSession()

r = session.get('https://www.dsebd.org/day_end_archive.php?startDate=2024-03-10&endDate=2024-03-10&inst=All%20Instrument&archive=data')
```

```python
r.html.find('.table.table-bordered.background-white.shares-table.fixedHeader>tbody')
```

```python
table = r.html.find('.table.table-bordered.background-white.shares-table.fixedHeader>tbody')[0].text.split('\n')
table = table[:-37] #Excluding last few anomalies
# table[-10:]
table[:20]
```

```python
date = []
trading_code = []
ltp = []
high = []
low = []
openp = []
closep = []
ycp = []
trade = []
value = []
volumn = []


for i in range(1, len(table), 12):
    date.append(table[i])
    trading_code.append(table[i+1])
    ltp.append(table[i+2])
    high.append(table[i+3])
    low.append(table[i+4])
    openp.append(table[i+5])
    closep.append(table[i+6])
    ycp.append(table[i+7])
    trade.append(table[i+8])
    value.append(table[i+9])
    volumn.append(table[i+10])
    
df_list = [date, trading_code, ltp, high, low, openp, closep, ycp, trade, value, volumn]
df_column = ['Date', 'Trading Code', 'LTP', 'High', 'Low', 'OpenP', 'CloseP', 'YCP', 'Trade', 'Value', 'Volumn']

df = pd.DataFrame(df_list).T
df.columns = df_column
df
```

```python
df.info()
```

```python
columns = ['LTP', 'High', 'Low', 'OpenP', 'CloseP', 'YCP', 'Trade', 'Value', 'Volumn']

for col in columns:
    df[col] = df[col].str.replace(',', '').astype('float64')
```

```python
df.info()
```

### Top Ten Gainer Considering CloseP & YCP on a Day

`%Change = ((CloseP- YCP) * 100) / YCP`

```python
df['%Change'] = round(((df['CloseP']-df['YCP'])*100)/df['YCP'], 4)
df
```

```python
df.sort_values(by=['%Change'], ascending=False, inplace=True)
df
```

```python
top_10_gainer_closeP_YCP = df.iloc[:10][['Trading Code', 'CloseP', 'High', 'Low', 'YCP', '%Change']]
top_10_gainer_closeP_YCP
```

```python
top_10_gainer_closeP_YCP.to_csv('Top 10 Gainer (CloseP and YCP).csv', index=False)
```

### Top Ten Loser Considering CloseP & YCP on a Day

```python
top_10_loser_closeP_YCP = df.iloc[-10:][['Trading Code', 'CloseP', 'High', 'Low', 'YCP', '%Change']].sort_values(by='%Change')
top_10_loser_closeP_YCP
```

```python
top_10_gainer_openP_LTP.to_csv('Top 10 Gainer (openP and LTP).csv', index=False)
```
